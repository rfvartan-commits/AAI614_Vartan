```python
# keep only numeric columns
numeric_grads = grads.select_dtypes(include=['number'])

Q1 = numeric_grads.quantile(0.25)
Q3 = numeric_grads.quantile(0.75)
IQR = Q3 - Q1

outliers = (numeric_grads < (Q1 - 1.5 * IQR)) | (numeric_grads > (Q3 + 1.5 * IQR))

outliers
```

Question 1: Identify all the outliers in the above data.  Justify your answers using objective measures.

One outlier was identified in the dataset. It appears in the “Avg Hours Studies per Week” column at row index 6.
This value is considered an outlier because it falls outside the expected statistical range based on the IQR method (below Q1 − 1.5×IQR or above Q3 + 1.5×IQR).

No outliers were found in the GPA column or the Sense of Humour column. All their values lie within the normal range.
Therefore, only one observation in the dataset is classified as an outlier, while all other data points are considered normal.

```python
grads.isnull().sum()
```

```python
grads['GPA'] = grads['GPA'].fillna(grads['GPA'].mean())
grads['Sense of Humour (0-5)'] = grads['Sense of Humour (0-5)'].fillna(grads['Sense of Humour (0-5)'].mean())
```
