## Random Walk

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import datetime
import random
import re

random.seed(datetime.datetime.now().timestamp())

def getLinks(articleUrl):
    url = Request(
        f'https://en.wikipedia.org{articleUrl}',
        headers={'User-Agent': 'Mozilla/5.0'}
    )
    html = urlopen(url)
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all(
        'a', href=re.compile('^(/wiki/)((?!:).)*$')
    )

links = getLinks('/wiki/Kevin_Bacon')

for _ in range(20):  # prevents infinite loop
    if not links:
        break
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```

## Recursively crawling an entire site

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import re

pages = set()

max_pages = 50
count = 0

def getLinks(pageUrl, depth=0):
    global count

    if depth > 25:
        return
    if count >= max_pages:
        return

    try:
        url = Request(
            f'https://en.wikipedia.org{pageUrl}',
            headers={'User-Agent': 'Mozilla/5.0'}
        )

        html = urlopen(url)
        bs = BeautifulSoup(html, 'html.parser')

    except:
        return

    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            newPage = link.attrs['href']

            if newPage not in pages:
                pages.add(newPage)
                count += 1

                print(f"{count}. {newPage}")

                if count >= max_pages:
                    return

                getLinks(newPage, depth + 1)

getLinks('')
```

## Collecting Data Across an Entire Site

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import re

pages = set()
max_pages = 50
count = 0

def getLinks(pageUrl, depth=0):
    global count

    if depth > 10:
        return
    if count >= max_pages:
        return

    try:
        url = Request(
            f'https://en.wikipedia.org{pageUrl}',
            headers={'User-Agent': 'Mozilla/5.0'}
        )

        html = urlopen(url)
        bs = BeautifulSoup(html, 'html.parser')

        print(bs.h1.get_text() if bs.h1 else "No title")

        bodyContent = bs.find('div', {'id':'bodyContent'})
        paragraphs = bodyContent.find_all('p') if bodyContent else []

        if paragraphs:
            print(paragraphs[0].get_text())

        edit = bs.find(id='ca-edit')
        if edit and edit.find('a'):
            print(edit.find('a').attrs.get('href'))

    except:
        return

    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            newPage = link.attrs['href']

            if newPage not in pages:
                pages.add(newPage)

                count += 1
                print('-' * 20)
                print(newPage)

                if count >= max_pages:
                    return

                getLinks(newPage, depth + 1)

getLinks('/wiki/General-purpose_programming_language')
```

## Crawling across the Internet

```python
from urllib.request import urlopen, Request
from urllib.parse import urlparse, urljoin
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup
import random

def getInternalLinks(bs, url):
    parsed_base = urlparse(url)
    internalLinks = set()

    for link in bs.find_all('a'):
        href = link.get('href')
        if not href:
            continue

        absolute = urljoin(url, href)
        parsed = urlparse(absolute)

        if parsed.netloc == parsed_base.netloc:
            internalLinks.add(absolute)

    return list(internalLinks)

def getExternalLinks(bs, url):
    parsed_base = urlparse(url)
    externalLinks = set()

    for link in bs.find_all('a'):
        href = link.get('href')
        if not href:
            continue

        absolute = urljoin(url, href)
        parsed = urlparse(absolute)

        if parsed.netloc and parsed.netloc != parsed_base.netloc:
            externalLinks.add(absolute)

    return list(externalLinks)

def getRandomExternalLink(startingPage):
    try:
        req = Request(startingPage, headers={'User-Agent': 'Mozilla/5.0'})
        html = urlopen(req, timeout=5)
        bs = BeautifulSoup(html, 'html.parser')
    except (HTTPError, URLError):
        return None

    externalLinks = getExternalLinks(bs, startingPage)

    if externalLinks:
        return random.choice(externalLinks)

    internalLinks = getInternalLinks(bs, startingPage)

    if internalLinks:
        return getRandomExternalLink(random.choice(internalLinks))

    return None

def followExternalOnly(startingSite, limit=10):
    current = startingSite

    for _ in range(limit):
        externalLink = getRandomExternalLink(current)

        if not externalLink:
            print("No external links found.")
            return

        print(f'Random external link is: {externalLink}')
        current = externalLink


followExternalOnly('https://en.wikipedia.org/wiki/Kevin_Bacon')
```

## Collect all External Links from a Site

```python
from urllib.request import urlopen, Request
from urllib.parse import urlparse, urljoin
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup
import re

allExtLinks = []
allIntLinks = []
visited = set()
max_pages = 50
count = 0

def getInternalLinks(bs, url):
    base_netloc = urlparse(url).netloc
    internalLinks = set()

    for link in bs.find_all('a'):
        href = link.get('href')
        if not href:
            continue

        absolute = urljoin(url, href)
        parsed = urlparse(absolute)

        if parsed.netloc == base_netloc:
            internalLinks.add(absolute)

    return list(internalLinks)

def getExternalLinks(bs, url):
    base_netloc = urlparse(url).netloc
    externalLinks = set()

    for link in bs.find_all('a'):
        href = link.get('href')
        if not href:
            continue

        absolute = urljoin(url, href)
        parsed = urlparse(absolute)

        if parsed.netloc and parsed.netloc != base_netloc:
            externalLinks.add(absolute)

    return list(externalLinks)

def getAllExternalLinks(url, depth=0):
    global count

    if depth > 10:
        return
    if count >= max_pages:
        return
    if url in visited:
        return

    visited.add(url)

    try:
        req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        html = urlopen(req, timeout=5)
        bs = BeautifulSoup(html, 'html.parser')
    except (HTTPError, URLError):
        return

    internalLinks = getInternalLinks(bs, url)
    externalLinks = getExternalLinks(bs, url)

    for link in externalLinks:
        if link not in allExtLinks:
            allExtLinks.append(link)
            print(link)

            count += 1
            if count >= max_pages:
                return

    for link in internalLinks:
        if link not in allIntLinks:
            allIntLinks.append(link)
            getAllExternalLinks(link, depth + 1)

getAllExternalLinks('https://en.wikipedia.org/wiki/Kevin_Bacon')
```
