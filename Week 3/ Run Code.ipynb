## Random Walk

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

random.seed(datetime.datetime.now().timestamp())

def getLinks(articleUrl):
    url = f'https://en.wikipedia.org{articleUrl}'

    headers = {'User-Agent': 'Mozilla/5.0'}
    req = Request(url, headers=headers)
    html = urlopen(req)

    bs = BeautifulSoup(html, 'html.parser')

    return bs.find('div', {'id': 'bodyContent'}).find_all(
        'a', href=re.compile('^(/wiki/)((?!:).)*$')
    )

visited = set()
max_pages = 10

links = getLinks('/wiki/Kevin_Bacon')

while len(links) > 0 and len(visited) < max_pages:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']

    if newArticle not in visited:
        visited.add(newArticle)
        print(newArticle)
        links = getLinks(newArticle)
```

## Recursively crawling an entire site

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import re
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

pages = set()

def getLinks(pageUrl):
    url = f'https://en.wikipedia.org{pageUrl}'

    headers = {'User-Agent': 'Mozilla/5.0'}
    req = Request(url, headers=headers)
    html = urlopen(req)

    bs = BeautifulSoup(html, 'html.parser')

    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            newPage = link.attrs['href']

            if newPage not in pages:
                print(newPage)
                pages.add(newPage)

                # safety limit (prevents infinite recursion crash)
                if len(pages) < 20:
                    getLinks(newPage)

getLinks('/wiki/Kevin_Bacon')
```

## Collecting Data Across an Entire Site

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import re
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

pages = set()

def getLinks(pageUrl):
    url = f'https://en.wikipedia.org{pageUrl}'

    headers = {'User-Agent': 'Mozilla/5.0'}
    req = Request(url, headers=headers)
    html = urlopen(req)

    bs = BeautifulSoup(html, 'html.parser')

    try:
        print(bs.h1.get_text())

        bodyContent = bs.find('div', {'id': 'bodyContent'}).find_all('p')
        if len(bodyContent) > 0:
            print(bodyContent[0].get_text())

        editLink = bs.find(id='ca-edit')
        if editLink:
            print(editLink.find('a').attrs['href'])

    except AttributeError:
        print('This page is missing something! Continuing.')

    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            newPage = link.attrs['href']

            if newPage not in pages:
                print('-' * 20)
                print(newPage)

                pages.add(newPage)

                # safety stop (VERY IMPORTANT)
                if len(pages) < 20:
                    getLinks(newPage)

getLinks('/wiki/General-purpose_programming_language')
```

## Crawling across the Internet

```python
from urllib.request import Request, urlopen
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup
import random
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

def open_url(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    req = Request(url, headers=headers)
    return urlopen(req)

def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    internalLinks = set()

    for link in bs.find_all('a', href=True):
        href = link['href']

        if href.startswith('#') or href.startswith('mailto:') or href.startswith('javascript:'):
            continue

        full_url = urljoin(url, href)

        if urlparse(full_url).netloc == netloc:
            internalLinks.add(full_url)

    return list(internalLinks)

def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()

    for link in bs.find_all('a', href=True):
        href = link['href']

        if href.startswith('#') or href.startswith('mailto:') or href.startswith('javascript:'):
            continue

        full_url = urljoin(url, href)

        if urlparse(full_url).netloc and urlparse(full_url).netloc != netloc:
            externalLinks.add(full_url)

    return list(externalLinks)

def getRandomExternalLink(startingPage):
    try:
        html = open_url(startingPage)
        bs = BeautifulSoup(html, 'html.parser')

        externalLinks = getExternalLinks(bs, startingPage)

        if externalLinks:
            return random.choice(externalLinks)

        internalLinks = getInternalLinks(bs, startingPage)

        if internalLinks:
            return getRandomExternalLink(random.choice(internalLinks))

        return None

    except Exception:
        return None

def followExternalOnly(startingSite, max_steps=10):
    current = startingSite
    visited = set()
    steps = 0

    while steps < max_steps:
        link = getRandomExternalLink(current)

        if not link:
            return

        if link not in visited:
            visited.add(link)
            print(link)
            current = link
            steps += 1

followExternalOnly("https://example.com/")
```

## Collect all External Links from a Site

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

allExtLinks = []
allIntLinks = []

def open_url(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    req = Request(url, headers=headers)
    return urlopen(req)

def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    links = set()

    for link in bs.find_all('a', href=True):
        href = link['href']

        if href.startswith('#') or href.startswith('mailto:') or href.startswith('javascript:'):
            continue

        full_url = urljoin(url, href)

        if urlparse(full_url).netloc == netloc:
            links.add(full_url)

    return list(links)

def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    links = set()

    for link in bs.find_all('a', href=True):
        href = link['href']

        if href.startswith('#') or href.startswith('mailto:') or href.startswith('javascript:'):
            continue

        full_url = urljoin(url, href)

        if urlparse(full_url).netloc and urlparse(full_url).netloc != netloc:
            links.add(full_url)

    return list(links)

def getAllExternalLinks(url, max_depth=10):
    if max_depth == 0:
        return

    try:
        bs = BeautifulSoup(open_url(url), 'html.parser')

        internalLinks = getInternalLinks(bs, url)
        externalLinks = getExternalLinks(bs, url)

        for link in externalLinks:
            if link not in allExtLinks:
                allExtLinks.append(link)
                print(link)

        for link in internalLinks:
            if link not in allIntLinks:
                allIntLinks.append(link)
                getAllExternalLinks(link, max_depth - 1)

    except Exception as e:
        print("Error:", e)

getAllExternalLinks("https://example.com")
```
